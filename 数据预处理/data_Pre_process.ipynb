```python
# 获取每个特征的离群点数量
outliers_count = {}
for feature in num_columns.columns:
    # 计算离群点的数量
    q1 = num_columns[feature].quantile(0.25)
    q3 = num_columns[feature].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers_count[feature] = len(num_columns[(num_columns[feature] < lower_bound) | (num_columns[feature] > upper_bound)])

# 输出每个特征的离群点数量
for feature, count in outliers_count.items():
    print(f"{feature}: {count} outliers")
```
